```python
def collatz(n):
    """ A good example of recursion."""
    
    if n == 1:
        return 
    
    elif n % 2 == 0:
        
        new_n = int(n/2)
        print(new_n)
    
        collatz(new_n)
        
    else:
        new_n = n*3 + 1
        print(new_n)
        
        collatz(new_n)
```
